```python
import pandas as pd

def drop_columns_with_many_nulls(df, threshold=800):
    """
    Supprime les colonnes d'un DataFrame qui ont plus de `threshold` valeurs nulles.
    
    Parameters:
        df (pd.DataFrame): le DataFrame à traiter
        threshold (int): le nombre de nulls au-delà duquel la colonne sera supprimée
        
    Returns:
        pd.DataFrame: DataFrame avec les colonnes supprimées
    """
    # Identifier les colonnes à supprimer
    
    cols_to_drop = df.columns[df.isna().sum() > threshold]
    print(f"Colonnes supprimées ({len(cols_to_drop)}): {list(cols_to_drop)}")
    
    # Supprimer les colonnes
    df_cleaned = df.drop(columns=cols_to_drop)
    return df_cleaned
```

```python
df = drop_columns_with_many_nulls(df)
```

```python
def show_example_non_nulls(df):
    """
    Affiche une valeur non nulle pour chaque colonne du DataFrame.
    """
    examples = {}
    for col in df.columns:
        # Sélectionner la première valeur non nulle
        non_null_value = df[col].dropna().iloc[0] if not df[col].dropna().empty else None
        examples[col] = non_null_value
    return pd.Series(examples)

# Exemple d'utilisation
example_values = show_example_non_nulls(df)
print(example_values)
```

```python
# Séparer la colonne 'ACR' en plusieurs colonnes
acr_split = df['ACR'].str.split(',', expand=True)

# Renommer les colonnes : acr1, acr2, acr3...
acr_split.columns = [f'ACR{i+1}' for i in range(acr_split.shape[1])]

# Convertir les valeurs en float
acr_split = acr_split.apply(pd.to_numeric, errors='coerce')

# Concaténer avec le DataFrame original si nécessaire
df = pd.concat([df.drop(columns=['ACR']), acr_split], axis=1)
```

```python
df.info()
```

```python
# Liste des colonnes à traiter
cols_to_fill = ["ACR1", "ACR2", "ACR3", "ACR4"]

# Remplacer les NaN par 0
df[cols_to_fill] = df[cols_to_fill].fillna(0)

# Vérifier
print(df[cols_to_fill].info())
```

# Remplir les colonnes vides

```python
textual_cols = [
    "Description", "ClinicalPresentation", "Commentary", 
    "Chapter", "Hospital", "Department", "Language" ,"Diagnosis"
]
df[textual_cols] = df[textual_cols].fillna("Non précisé")
```

```python
numeric_cols = [ "ODislocation", "OPolytrauma", "OOpen", "OPathologic", "OGraft"]
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python
# ✅ Conversion avec gestion d’erreurs
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y', errors='coerce')
df['Birthdate'] = pd.to_datetime(df['Birthdate'], format='%d.%m.%Y', errors='coerce')
df['Creation'] = pd.to_datetime(df['Creation'], format='%d.%m.%Y', errors='coerce')
```
